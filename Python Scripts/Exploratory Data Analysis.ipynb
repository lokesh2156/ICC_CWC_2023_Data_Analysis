## Univariate Analysis

```python
numerical_columns = ['Runs', 'Balls', 'Fours', 'Sixes', 'StrikeRate']
plt.figure(figsize=(15, 8))
plt.suptitle('Batting Numerical Columns Analysis', fontsize=16)
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.histplot(bat_df[column], kde=True, color='skyblue', bins=20)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.suptitle('Batting Numerical Columns Analysis (Box Plots)', fontsize=16)

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=bat_df[column], color='salmon')
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  
plt.show()
```

```python
numerical_columns = ['Overs', 'Maidens', 'Runs', 'Wickets', 'No_Balls','Wides','Economy']
plt.figure(figsize=(15, 8))
plt.suptitle('Bowling Numerical Columns Analysis', fontsize=16)
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(bowl_df[column], kde=True, color='skyblue', bins=20)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
plt.show()
```

```python
numerical_columns = ['Overs', 'Maidens', 'Runs', 'Wickets', 'No_Balls','Wides','Economy']
plt.figure(figsize=(15, 8))
plt.suptitle('Bowling Numerical Columns Analysis (Box Plots)', fontsize=16)

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=bowl_df[column], color='salmon')
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  
plt.show()
```

```python
dismissal_counts = bat_df['Dismissal Type'].value_counts()

plt.figure(figsize=(12, 6))
plt.pie(dismissal_counts, labels=dismissal_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Dismissal Type Distribution ')

plt.tight_layout()
plt.show()
```

```python
# Bowler Distribution Type
plt.figure(figsize=(15, 5))
temp = bowl_df.groupby('Bowler_Type')['Bowler'].count()
ax = sns.barplot(x=temp.index, y=temp.values, palette='Set2')
total_bowlers = temp.sum()
percentage_values = (temp / total_bowlers) * 100
for i, value in enumerate(percentage_values):
    ax.text(i, temp.values[i], f' {value:0.01f}%', va='bottom', ha='center', fontsize=10)

plt.xlabel('Bowler Type')
plt.ylabel('Bowler Count')
plt.title('Bowler Type Distribution')
plt.tight_layout()
plt.show()
```

```python
# Highest Runs
highest_scores = bat_df.groupby('Batsman')['Runs'].max().sort_values(ascending=False)[:10]
plt.figure(figsize=(10, 6))
light_green = '#90EE90'
yellow = '#FFFF00'
sns.barplot(x=highest_scores.index, y=highest_scores.values, palette='muted')
for i, v in enumerate(highest_scores):
    plt.text(i, v + 10, str(highest_scores[i]), ha='center', va='bottom', fontsize=10)
plt.xticks(rotation=30)
plt.title('Top 10 Highgest Scores')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.tight_layout()
plt.show()
```

```python
#Most Centuries (100s)
centuries_df = bat_df.loc[(bat_df['Runs'] >= 100) & (bat_df['Runs'] < 200), ['Batsman', 'Runs', 'Team']]
top_centuries = centuries_df.groupby(['Batsman', 'Team'])['Runs'].count().sort_values(ascending=False)[:10].reset_index()

# Most Fifties (50s)
fifties_df = bat_df.loc[(bat_df['Runs'] >= 50) & (bat_df['Runs'] < 100), ['Batsman', 'Runs', 'Team']]
top_fifties = fifties_df.groupby(['Batsman', 'Team'])['Runs'].count().sort_values(ascending=False)[:10].reset_index()


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

sns.barplot(ax=axes[0], x='Batsman', y='Runs', data=top_centuries, palette='bright')
axes[0].set_title('Top 10 Batsmen with Most Centuries (100s)')
axes[0].set_xlabel('Batsman')
axes[0].set_ylabel('Number of Centuries')
axes[0].tick_params(axis='x', rotation=30)

sns.barplot(ax=axes[1], x='Batsman', y='Runs', data=top_fifties, palette='colorblind')
axes[1].set_title('Top 10 Batsmen with Most Fifties (50s)')
axes[1].set_xlabel('Batsman')
axes[1].set_ylabel('Number of Fifties')
axes[1].tick_params(axis='x', rotation=30)

for ax in axes:
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10, color='black',fontweight='bold')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.tight_layout()
plt.show()
```

```python
## Team wise Most Fifties & Tons 

# Most Centuries (100s) by Team
centuries_df = bat_df.loc[(bat_df['Runs'] >= 100) & (bat_df['Runs'] < 200), ['Batsman', 'Runs', 'Team']]
top_centuries = centuries_df.groupby(['Team'])['Runs'].count().sort_values(ascending=False)[:10].reset_index()

#Most Fifties (50s) by Team
fifties_df = bat_df.loc[(bat_df['Runs'] >= 50) & (bat_df['Runs'] < 100), ['Batsman', 'Runs', 'Team']]
top_fifties = fifties_df.groupby(['Team'])['Runs'].count().sort_values(ascending=False)[:10].reset_index()
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

sns.barplot(ax=axes[0], x='Team', y='Runs', data=top_centuries, palette='colorblind')
axes[0].set_title('Team wise Centuries (100s)')
axes[0].set_xlabel('Team')
axes[0].set_ylabel('Number of Centuries')
axes[0].tick_params(axis='x', rotation=30)

sns.barplot(ax=axes[1], x='Team', y='Runs', data=top_fifties, palette='muted')
axes[1].set_title('Team wise Fifties (50s)')
axes[1].set_xlabel('Team')
axes[1].set_ylabel('Number of Fifties')
axes[1].tick_params(axis='x', rotation=30)

for ax in axes:
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=10, color='black',fontweight='bold')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.tight_layout()
plt.show()
```

### Strike Rate

```python
fig, axes = plt.subplots(1,2, figsize=(18, 8))

# Subplot 1: Highest Strike Rate By Batsman
min_balls = bat_df[bat_df['Balls'] > 10]['Balls']  # Considering Min 10 Balls for Strike Rate
stb = bat_df[(bat_df['Balls'].isin(min_balls))].sort_values(by='StrikeRate', ascending=False)[:10]
ax1 = sns.barplot(ax=axes[0], y=stb['Batsman'], x=stb['StrikeRate'], orient='h', palette='viridis')
for i, v in enumerate(stb['StrikeRate']):
    ax1.text(v, i, f' {v:.2f}', va='center', ha='left', fontsize=10, color='black')

axes[0].set_xlabel('Strike Rate')
axes[0].set_ylabel('Batsman')
axes[0].set_title('Top 10 Batsmen with Highest Strike Rate')

# Subplot 2: Average Strike Rate for Each Team
tsr = bat_df.groupby('Team')['StrikeRate'].mean().sort_values(ascending=False)
ax2 = sns.barplot(ax=axes[1], x=tsr.values, y=tsr.index, palette='muted')
for i, v in enumerate(tsr.values):
    ax2.text(v, i, f' {v:.2f}', va='center', ha='left', fontsize=10, color='black')

axes[1].set_xlabel('Strike Rate')
axes[1].set_ylabel('Team')
axes[1].set_title('Average Strike Rate for Each Team')

plt.tight_layout()
plt.show()
```
