```python
import pandas as pd 
url = 'https://en.wikipedia.org/wiki/List_of_current_Indian_chief_ministers'
data = pd.read_html(url)
data
```

```python
data[0]
```

```python
df = data[0]
df.head()
```

### Dropping unnecessary Columns

```python
df.drop(columns=['Portrait','Party[a].1','Alliance.1','Ministry','Ref'],inplace=True)
df.head()
```

### Renaming Columns 

```python
df.rename({'State (past chief ministers)':'State',
           'Name[4]':'Name',
           'Took office (tenure length)':'Tenure',
           'Party[a]':'Party'},inplace=True,axis=1)
```

```python
df.head()
```

```python
df['State'].replace(r'\(list\)', '', regex=True, inplace=True)
df.head()
```

```python
df.to_csv('F:\Freelance\CM_Wikipedia.csv',index=False)
```
